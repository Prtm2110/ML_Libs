# Relational Plots

Relational plots are used to explore relationships between variables.

## Scatter plot

Scatter plots are ideal for visualizing the relationship between two continuous variables. They help identify correlations, trends, and potential outliers.

Key Parameters:

data: The dataset to plot.

x, y: Variables to be plotted on the x and y axes.

hue:  Grouping variable that will produce points with different colors.

style: Grouping variable that will produce points with different markers.

size: Grouping variable that will produce points with different sizes.

```python
# Create a scatter plot
sns.scatterplot(data=df, x='total_bill', y='tip', hue='day', style='time', size='size')

# Add titles and labels
plt.title('Total Bill vs Tip by Day and Time')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')

# Display the plot
plt.show()
```

## Line plot

Line plots are useful for visualizing trends over a continuous interval or time period. They are commonly used to display changes in variables over time.

```python
# Calculate the average total bill for each party size and time
df_grouped = df.groupby(['size', 'time'])['total_bill'].mean().reset_index()

# Create a line plot
sns.lineplot(data=df_grouped, x='size', y='total_bill', hue='time', marker='o')

# Add titles and labels
plt.title('Average Total Bill by Party Size and Time')
plt.xlabel('Party Size')
plt.ylabel('Average Total Bill ($)')

# Display the plot
plt.show()
```

# Categorical Plots

Categorical plots are designed to visualize the distribution of categorical variables and their relationships with other variables

## Bar Plot

Bar plots display the relationship between a categorical variable and a continuous variable by showing the mean of the continuous variable for each category.


A confidence interval (CI) is a fundamental concept in statistics that provides a range of values used to estimate a population parameter (such as a mean or proportion) with a certain level of confidence. It offers a way to express the uncertainty inherent in using sample data to make inferences about a larger population.


```python
# Create a bar plot
sns.barplot(data=df, x='day', y='total_bill', palette='pastel')

# Add titles and labels
plt.title('Average Total Bill by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Average Total Bill ($)')

# Display the plot
plt.show()
```

## Count Plot

Count plots are used to display the count of observations in each categorical bin using bars.

```python
# Create a count plot
sns.countplot(data=df, x='sex', palette='muted')

# Add titles and labels
plt.title('Count of Tips by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')

# Display the plot
plt.show()
```

## Box Plot

Box plots summarize the distribution of a continuous variable through their quartiles and highlight potential outliers.

```python
# Create a box plot
sns.boxplot(data=df, x='day', y='total_bill', palette='Set2')

# Add titles and labels
plt.title('Total Bill Distribution by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')

# Display the plot
plt.show()
```

## Voilin plot

Violin plots combine the features of box plots and kernel density plots. They show the distribution of the data across different categories and provide insights into the probability density of the data at different values.

```python
# Create a violin plot
sns.violinplot(data=df, x='day', y='total_bill', palette='Set2')

# Add titles and labels
plt.title('Total Bill Distribution by Day and Gender')
plt.xlabel('Day of the Week')
plt.ylabel('Total Bill ($)')

# Display the plot
plt.show()
```

# Distribution Plots

Distribution plots help visualize the distribution of a single continuous variable or compare distributions across multiple categories.

## Histogram

Histograms display the distribution of a single continuous variable by dividing the data into bins and counting the number of observations in each bin.

Key Parameters:

data: The dataset to plot.

x: Variable to be plotted on the x-axis.

bins: Number of bins to divide the data.

kde: Whether to include a kernel density estimate.

```python
# Create a histogram with KDE
sns.histplot(data=df, x='total_bill', bins=20, kde=True, color='skyblue')

# Add titles and labels
plt.title('Distribution of Total Bills')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')

# Display the plot
plt.show()
```

## Kde plot

Kernel Density Estimation (KDE) plots provide a smooth estimate of the data distribution, useful for visualizing the probability density function of a continuous variable.

Key Parameters:

data: The dataset to plot.

x, y: Variables to be plotted on the x and y axes

hue: Grouping variable that will produce KDEs with different colors.

multiple: How to handle multiple densities ('layer', 'stack', 'fill').

```python
# Create a KDE plot
sns.kdeplot(data=df, x='total_bill', hue='sex', multiple='stack', palette='Set2')

# Add titles and labels
plt.title('KDE of Total Bills by Gender')
plt.xlabel('Total Bill ($)')
plt.ylabel('Density')

# Display the plot
plt.show()
```

# Advanced Plotting Techniques

While basic plots are essential for exploratory data analysis, advanced plotting techniques in Seaborn empower you to uncover deeper insights and present data in more nuanced ways. In this section, we'll explore advanced visualization tools such as Pair Plots, Joint Plots, Facet Grids, and Heatmaps.

## Joint plot

Joint plots combine a scatter plot (or other types of plots) with univariate plots (histograms or KDE plots) on the axes to provide a comprehensive view of the relationship between two variables along with their distributions. This dual visualization helps in understanding both the correlation and the distribution of the data.

**Key Parameters:**

data: The dataset to plot.

x, y: Variables to plot on the x and y axes.

hue: Grouping variable that adds color differentiation.

kind: Type of joint plot ('scatter', 'reg', 'kde', 'hex').

height: Size of the plot.

ratio: Ratio of joint to marginal axes.

**Use Cases:**

Correlation Analysis: Assess the strength and direction of relationships between two variables.

Regression Analysis: Visualize the fit of a regression model.

Distribution Comparison: Compare how variables are distributed individually and jointly.

```python
# Create a joint plot with regression
sns.jointplot(data=df, x='total_bill', y='tip', kind='reg', height=8, ratio=5, color='blue')

# Add a title
plt.suptitle('Joint Plot of Total Bill and Tip', y=1.02)
```

```python
# Create a joint plot with regression
sns.jointplot(data=df, x='total_bill', y='tip', kind='hex')

# Add a title
plt.suptitle('Joint Plot of Total Bill and Tip', y=1.02)
```

# Pair Plot

Pair plots are an excellent way to visualize pairwise relationships between multiple variables in a dataset. They provide a matrix of scatter plots for each pair of variables, along with histograms or KDE plots on the diagonal to show individual variable distributions.

**Key Parameters:**

data: The dataset to plot.

hue: Grouping variable that adds color differentiation.

palette: Color palette to use for different hues.

kind: Type of plot to draw ('scatter', 'reg').

diag_kind: Kind of plot for the diagonal ('hist', 'kde').

markers: Marker style for scatter plots.

```python
sns.pairplot(df, height=2.5)
```

```python
sns.pairplot(df,hue='sex')
```

 # **Heatmap (IMP)**

 Heatmaps are used to visualize data matrices, where individual values are represented as colors. They are particularly effective for displaying correlation matrices, confusion matrices, or any data where you need to visualize the intensity of values across two dimensions.

**Key Parameters:**

data: The dataset to plot

annot:  Whether to annotate the cells with their values.

fmt: String formatting code for annotations.

cmap:  Colormap to use.

linewidths: Width of the lines that will divide each cell.

linecolor: Color of the lines that will divide each cell.

Use Cases:

Correlation Analysis: Visualize relationships between multiple numerical variables.

Data Matrix Visualization: Display values in a matrix format with color coding.

Pattern Recognition: Identify clusters or patterns in data matrices.

```python
# Select numeric columns
tips_numeric = df.select_dtypes(include=['float64', 'int64'])

corr = tips_numeric.corr()

print(corr)
```

```python
sns.heatmap(corr)
```

```python

# Create a heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap = 'Purples', linewidths=0.5, linecolor='white')

plt.title('Correlation Matrix of Tips Dataset', fontsize=16)
plt.show()
```
