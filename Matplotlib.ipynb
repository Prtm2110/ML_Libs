# Matplotlib Library
Matplotlib is a comprehensive library used for creating static, interactive, and animated visualizations in Python. It is one of the most popular and powerful plotting libraries in Python, allowing you to create high-quality plots, charts, and figures. It is widely used in data science, machine learning, and scientific computing for visualizing data, trends, and relationships between variables.



## key features of Matplotlib


1.   #### Variety of Plots: It supports line plots, bar plots, scatter plots, histograms, pie charts, etc.
2.   #### Cutomization: Matplotlib allows you to customize almost every aspect of your plot (title, axes, grid, colors, markers, etc.).
3. #### Interactive Plotting : Matplotlib can create interactive plots when used with Jupyter Notebooks, PyCharm, or other interactive Python environments.

##Bar chart


1.   generally numerical vs categorical
2.   example:- aggregate analysis of groups



```python
# simple bar chart
children = [10,20,40,10,30]
colors = ['red','green','blue','yellow','pink']

plt.bar(colors,children)
```

```python
#horizontal bar chart
plt.barh(colors,children)
```

```python
# color and label
df3 = pd.read_csv('/content/batsman_season_record.csv')
```

```python
plt.bar(df3['batsman'],df3['2015']) #blue
plt.bar(df3['batsman'],df3['2016']) #orange
plt.bar(df3['batsman'],df3['2017']) #green
# graph getting stack on one another wrong way
```

```python
plt.bar(np.arange(df3.shape[0])-0.2,df3['2015'], width=0.2 ) #blue
plt.bar(np.arange(df3.shape[0]),df3['2016'], width=0.2) #orange
plt.bar(np.arange(df3.shape[0])+0.2,df3['2017'], width=0.2, color='red') #green->red

plt.xticks(np.arange(df3.shape[0]),df3['batsman'])

plt.show()
```

```python
#problem name is very big
children = [10,20,40,10,30]
colors = ['red red red red red ','green green green green','blue blue blue blue blue','yellow yellow yellow yellow yellow','pink pink pink pink pink']

plt.bar(colors,children)
plt.xticks(rotation=90)
```

```python
# stacked bar chart
plt.bar(df3['batsman'],df3['2017'],label='2017')
plt.bar(df3['batsman'],df3['2016'],bottom=df3['2017'],label='2016')
plt.bar(df3['batsman'],df3['2015'],bottom=(df3['2016']+df3['2017']),label='2015')

plt.legend()
plt.show()
```

## HeatMap

```python
df5 = pd.read_csv('/content/IPL_Ball_by_Ball_2008_2022.csv')
df5.head()
```

```python
# in data when i checked there is anamoly that we have balls more than 6 so to handle that and only taking the value of the balls in which batter scored 6 we used this
temp_df= df5[(df5['ballnumber'].isin([1,2,3,4,5,6])) & (df5['batsman_run']==6)]
```

```python
temp_df.head()
```

```python
grid = temp_df.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
grid
```

```python
plt.figure(figsize=(20,10))
plt.imshow(grid)
plt.colorbar()
```
